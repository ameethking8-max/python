2. What are the key assumptions of Simple Linear Regression?
 - The key assumptions of Simple Linear Regression ensure that the model is valid and its results are reliable. These assumptions relate to the relationship between variables and the behavior of errors.

- Linearity
  * The relationship between the independent variable
  * x and the dependent variable y is linear.
    * The mean of ùë¶ changes linearly with x.
    
  * Independence of Errors
  * The residuals (errors) are independent of each other.
    * Commonly violated in time-series data.  

  * Homoscedasticity (Constant Variance)
  * The variance of the residuals is constant across all values of x.
    * Residual spread should be roughly the same along the regression line.

  * Normality of Errors
  * The residuals are normally distributed.
    * Mainly important for hypothesis testing and confidence intervals.   

  * No Measurement Error in x (often implied)
  * The independent variable is measured without error.

6. Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
- X = [1, 2, 3, 4, 5]
- Y = [2.1, 4.3, 6.1, 7.9, 10.2]
- Plot the regression line over the data points.  

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and train the model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Print model parameters
print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)

# Plot data points
plt.scatter(X, Y, color='blue', label='Data Points')

# Plot regression line
plt.plot(X, Y_pred, color='red', label='Regression Line')

# Labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()

# Show plot
plt.show()
```

8.  Implement polynomial regression on the following data:
- X = [1, 2, 3, 4, 5]
- Y = [2.2, 4.8, 7.5, 11.2, 14.7]
- Fit a 2nd-degree polynomial and plot the resulting curve.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial features (degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict values
Y_pred = model.predict(X_poly)

# Plot original data points
plt.scatter(X, Y, color='blue', label='Data Points')

# Plot polynomial regression curve
plt.plot(X, Y_pred, color='red', label='2nd Degree Polynomial Fit')

# Labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression (Degree 2)')
plt.legend()

# Show plot
plt.show()

# Print model coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

9.  Create a residuals plot for a regression model trained on this data:
- X = [10, 20, 30, 40, 50]
- Y = [15, 35, 40, 50, 65]
- Assess heteroscedasticity by examining the spread of residuals

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Convert to numpy arrays and reshape
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit simple linear regression
model = LinearRegression()
model.fit(X, Y)

# Predictions
Y_pred = model.predict(X)

# Residuals
residuals = Y - Y_pred

# Plot residuals
plt.scatter(X, residuals, color='purple')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.show()
```

- Scenario
  * Predicting house prices using Area, Number of Rooms, Location
  * Detected issues: heteroscedasticity and multicollinearity

- Diagnose the Problems
- Heteroscedasticity
  * Occurs when the variance of residuals is not constant (errors increase with the value of predictors or target).
  * Detect using:
     * Residuals vs fitted values plot
     * Breusch-Pagan test or White test  

- Multicollinearity
  * Occurs when predictors are highly correlated with each other.
  * Detect using:
     * Variance Inflation Factor (VIF)
     * Correlation matrix

- Address Multicollinearity
  * Remove or combine correlated features     
     * Example: If Area and Number of Rooms are highly correlated, consider combining them into a single metric like area per room.
  * Feature selection
     * Use stepwise regression, LASSO, or tree-based models to reduce redundant variables.
  * Dimensionality reduction
     * Apply Principal Component Analysis (PCA) to transform correlated features into uncorrelated components.    

- Address Heteroscedasticity
  * Transform the dependent variable
     * Use logarithm, square root, or Box-Cox transformation on house prices to stabilize variance.
  * Weighted Least Squares (WLS)
     * Assign weights inversely proportional to variance of residuals to give less importance to high-variance points.
  * Use robust regression methods
     * Methods like Huber Regression or RANSAC are less sensitive to heteroscedasticity.     

- Model Evaluation & Validation
  * Cross-validation
     * Ensure that model performance is consistent across subsets of data.
  * Check residuals again
    * After corrections, plot residuals to ensure variance is more constant.
  * Compare model performance
    * Use metrics like RMSE, MAE, R¬≤ to confirm improvements.     

- Consider Alternative Models
  * If problems persist, consider tree-based models like:
     * Random Forest Regressor
     * Gradient Boosting Machines
  * These models are robust to multicollinearity and heteroscedasticity without requiring transformations.
