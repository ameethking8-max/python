1. What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?

  - In hypothesis testing, we make decisions based on sample data, but because we‚Äôre dealing with probability, there‚Äôs always a chance of making an error. Type I and Type II errors describe the two ways our conclusions can be wrong.

   * Type I Error (False Positive)
       * Definition: Rejecting the null hypothesis H0 when it is actually true.
       * Analogy: A ‚Äúfalse alarm.‚Äù
       * Symbol: ùõº Œ± (alpha), the significance level‚Äîcommonly 0.05.
       * Example: Concluding a new drug works when it actually has no real effect.

   * Impact on decision-making:
       * Setting a lower ùõº Œ± reduces the chance of a Type I error, making you more cautious about claiming an effect.

       * But lowering ùõº Œ± makes it harder to detect real effects, which increases the chance of a Type II error.   

  * Type II Error (False Negative)  
      * Definition: Failing to reject H0 when the alternative hypothesis H1 is actually true.
      * Analogy: Missing a real effect.
      * Symbol:ùõΩ(beta).
      * Related Concept: Power of a test =1‚àíŒ≤.
      * Example: Concluding a medical test is ineffective when it actually does work.

   * Impact on decision-making:
      * High Œ≤ means you're missing real signals.   
      * Reducing Œ≤ (increasing power) usually requires:
        * larger sample size,
        * stronger effects, or
        * higher Œ± (being more willing to risk Type errors).

  * Trade-off Between Type I and Type II Errors   
      * Reducing one generally increases the other.
        * Strict criteria (low Œ±) ‚Üí fewer false positives but more false negatives.
        * Lenient criteria (high Œ±) ‚Üí fewer missed effects but more false alarms.

2. What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?

  * What Is the P-Value?
     * The P-value is the probability of obtaining a test statistic at least as extreme as the one observed, assuming the null hypothesis ùêª0 is true.
     * ‚ÄúIf the null hypothesis were true, how likely is it that we would see results this unusual?‚Äù

  * How to Interpret the P-Value
    * Small P-value ‚Üí Evidence against the null hypothesis  
       * A small P-value means the observed data would be very unlikely if H0 were true.
       * So we have reason to reject H0
       * If P ‚â§ Œ± (e.g., 0.05) ‚Üí reject H0

  * Large P-value ‚Üí Not enough evidence to reject the null hypothesis
    * A large P-value means the data are consistent with what we‚Äôd expect under H0
    * So we fail to reject H0

3. Explain the difference between a Z-test and a T-test, including when to use
each.

  * Z-Test vs. T-Test         
    * Z-Test
      * The population standard deviation (œÉ) is known, and
      * The sample size is large (typically n‚â•30), or the data are known to be normally distributed.

  * Common Uses
    * Testing means with known population variance
    * Proportion tests (large samples)
    * Quality control and industrial settings   

  * Distribution Used
    * Uses the standard normal distribution (Z-distribution)

  * T-Test    
    * The population standard deviation (œÉ) is unknown, and
    * The sample size is small (n< 30), or
    * You are estimating the mean from a sample.

  * Types of T-Tests  
    * One-sample t-test: Compare sample mean to a known value
    * Two-sample t-test: Compare means of two independent groups
    * Paired t-test: Compare means before/after or matched pairs

 * Distribution Used
   * Uses the t-distribution, which has heavier tails than the normal distribution
   * The shape depends on degrees of freedom (df); as sample size increases, the t-distribution approaches the normal distribution.   

4. What is a confidence interval, and how does the margin of error influence
its width and interpretation?

  * What Is a Confidence Interval?
    * A confidence interval has the general form:
      * Estimate¬±Margin of Error
    * Example: A 95% CI for a population mean might look like:
      * 65¬±3(i.e., from 62 to 68)   

  * What Is the Margin of Error?
    * The margin of error (MOE) reflects the amount of uncertainty in the estimate.It depends on:
      * Critical value (e.g., 1.96 for 95% CI using Z-distribution)
      * Standard error (which depends on sample size and variability)
      * Margin of Error=(critical value)√ó(standard error)   

 * How the Margin of Error Influences the CI
   * Larger Margin of Error ‚Üí Wider Confidence Interval
     * More uncertainty in the estimate
     * Indicates less precision
     * Could be caused by small sample sizes, high variability, or high confidence levels (e.g., 99% CI)

 * Smaller Margin of Error ‚Üí Narrower Confidence Interval
    * Greater precision
    * Caused by large sample sizes, low variability, or lower confidence levels (e.g., 90% CI)    

 * Factors That Increase the Margin of Error
    * Higher confidence level (95% ‚Üí 99% increases MOE)
    * Smaller sample size
    * Larger standard deviation in the data  

 * Factors That Decrease the Margin of Error
   * Lower confidence level (95% ‚Üí 90%)
   * Larger sample size
   * Lower variability   

5. Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?
  
  * What Is ANOVA?
    * ANOVA (Analysis of Variance) is a statistical test used to determine whether there are significant differences among the means of three or more groups.

  * Purpose of ANOVA
    * To test the null hypothesis that all group means are equal:
      * H0:Œº1=Œº2=Œº3=‚ãØ
    * Against the alternative hypothesis that at least one mean differs:
       * Ha:At least one group mean is different
    * ANOVA compares variability between groups to variability within groups using an F-statistic.

  * Why Use ANOVA Instead of Multiple t-Tests?
    * If you compared each pair of groups with separate t-tests:
       * You increase the probability of Type I errors (false positives).
       * ANOVA controls the overall error rate by testing all groups simultaneously in one model.
    * Thus, ANOVA is the standard method for comparing many groups reliably  
   
  * Assumptions of ANOVA
    * ANOVA relies on three key assumptions:

  * Independence of observations
    * Data in each group must be collected independently.
    * This is typically ensured through study design (e.g., random sampling, random assignment).  

  * Normality of residuals  
    * The data within each group should be approximately normally distributed.
    * ANOVA is fairly robust to moderate violations, especially with large sample sizes.

  * Homogeneity of variances (homoscedasticity)  
    * All groups should have roughly equal variances.
    * This is tested with methods like Levene‚Äôs test or Bartlett‚Äôs test.

  * How ANOVA Extends Hypothesis Testing Beyond Two Groups
    * With two groups, you would typically use a t-test to compare means.
    * With three or more groups, ANOVA:
       * Computes the variance between groups
       * Computes the variance within groups
       * Forms an F-ratio:
       * ùêπ=Between-group variance / Within-group variance
       * If the F-value is large, it suggests groups are different.  

  * After ANOVA: Post-Hoc Tests
    * If ANOVA is significant, it only tells you that at least one group differs, not which ones.
    * Post-hoc tests (e.g., Tukey, Bonferroni, Scheff√©) identify the specific pairs of groups that differ.     

6. Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset.    

```python
import numpy as np
from scipy.stats import norm

# -----------------------------
# Example Data and Parameters
# -----------------------------
data = [52, 49, 50, 51, 53, 48, 54, 50, 49, 52]  # sample data
mu_0 = 50             # hypothesized population mean (null hypothesis)
sigma = 3             # known population standard deviation
alpha = 0.05          # significance level

# -----------------------------
# Step 1: Compute sample mean
# -----------------------------
sample_mean = np.mean(data)
n = len(data)

# -----------------------------
# Step 2: Compute Z-statistic
# -----------------------------
z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# -----------------------------
# Step 3: Compute p-value (two-tailed)
# -----------------------------
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# -----------------------------
# Step 4: Interpret the result
# -----------------------------
print("Sample Mean:", round(sample_mean, 3))
print("Z-Statistic:", round(z_stat, 3))
print("P-Value:", round(p_value, 4))

if p_value < alpha:
    print(f"\nDecision: Reject H0 at Œ± = {alpha}")
    print("Interpretation: There is significant evidence that the true mean is different from", mu_0)
else:
    print(f"\nDecision: Fail to reject H0 at Œ± = {alpha}")
    print("Interpretation: There is NOT enough evidence to conclude that the true mean differs from", mu_0)
```

7. Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using
NumPy and plot the histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Step 1: Simulate binomial data
# -----------------------------
n_trials = 10     # number of trials per experiment
p_success = 0.5   # probability of success
sample_size = 1000  # number of experiments (samples)

# Simulate the dataset
data = np.random.binomial(n=n_trials, p=p_success, size=sample_size)

# -----------------------------
# Step 2: Plot histogram
# -----------------------------
plt.figure(figsize=(8,5))
plt.hist(data, bins=range(n_trials+2), edgecolor='black', alpha=0.7)
plt.title('Histogram of Binomially Distributed Data (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(range(n_trials+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

8. Generate multiple samples from a non-normal distribution and implement
the Central Limit Theorem using Python.


```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Step 1: Parameters
# -----------------------------
population_size = 100000  # large population
sample_size = 50          # size of each sample
num_samples = 1000        # number of samples

# -----------------------------
# Step 2: Generate non-normal population
# Exponential distribution (skewed)
# -----------------------------
population = np.random.exponential(scale=2.0, size=population_size)

# -----------------------------
# Step 3: Take multiple samples and compute sample means
# -----------------------------
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

# -----------------------------
# Step 4: Plot
# -----------------------------
plt.figure(figsize=(12,5))

# Histogram of original population
plt.subplot(1,2,1)
plt.hist(population, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Population (Exponential)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Histogram of sample means
plt.subplot(1,2,2)
plt.hist(sample_means, bins=30, color='salmon', edgecolor='black', alpha=0.7)
plt.title(f'Histogram of Sample Means (n={sample_size}, {num_samples} samples)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

9. Write a Python function to calculate and visualize the confidence interval
for a sample mean.


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def confidence_interval_visual(sample, confidence=0.95):
    """
    Calculates and visualizes the confidence interval for a sample mean.

    Parameters:
    - sample: list or numpy array of sample data
    - confidence: confidence level (default 0.95)
    """
    # Convert to numpy array
    sample = np.array(sample)
    n = len(sample)
    mean = np.mean(sample)
    std_err = stats.sem(sample)  # standard error = s / sqrt(n)

    # t-critical value for two-tailed test
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Margin of error
    margin_error = t_crit * std_err

    # Confidence interval
    ci_lower = mean - margin_error
    ci_upper = mean + margin_error

    print(f"Sample Mean: {mean:.3f}")
    print(f"{confidence*100:.1f}% Confidence Interval: ({ci_lower:.3f}, {ci_upper:.3f})")

    # -----------------------------
    # Visualization
    # -----------------------------
    plt.figure(figsize=(6,5))
    plt.bar(1, mean, yerr=margin_error, capsize=10, color='skyblue', edgecolor='black')
    plt.xlim(0,2)
    plt.xticks([1], ['Sample Mean'])
    plt.ylabel('Value')
    plt.title(f'{int(confidence*100)}% Confidence Interval')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()


# -----------------------------
# Example Usage
# -----------------------------
sample_data = [52, 49, 50, 51, 53, 48, 54, 50, 49, 52]
confidence_interval_visual(sample_data, confidence=0.95)
```

10. Perform a Chi-square goodness-of-fit test using Python to compare
observed and expected distributions, and explain the outcome.


```python
import numpy as np
from scipy.stats import chisquare

# -----------------------------
# Step 1: Define observed and expected frequencies
# -----------------------------
observed = np.array([50, 30, 20])   # observed counts
expected = np.array([40, 40, 20])   # expected counts under H0

# -----------------------------
# Step 2: Perform Chi-square goodness-of-fit test
# -----------------------------
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# -----------------------------
# Step 3: Interpret the result
# -----------------------------
alpha = 0.05  # significance level

print(f"Chi-square Statistic: {chi2_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print(f"\nDecision: Reject H0 at Œ± = {alpha}")
    print("Interpretation: The observed distribution significantly differs from the expected distribution.")
else:
    print(f"\nDecision: Fail to reject H0 at Œ± = {alpha}")
    print("Interpretation: The observed distribution does not significantly differ from the expected distribution.")
```
