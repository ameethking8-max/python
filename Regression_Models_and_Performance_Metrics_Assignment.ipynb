1. What is Simple Linear Regression (SLR)? Explain its purpose
- Simple Linear Regression (SLR) is a statistical method used to model and analyze the relationship between one independent variable (X) and one dependent variable (Y) by fitting a straight line to observed data.
- Mathematical form

- Purpose of Simple Linear Regression
- The main purposes of SLR are:
  * To understand relationships
    * It helps determine whether and how strongly the independent variable X affects the dependent variable Y

  * To quantify the effect
    * The slope (ğ›½1) measures how much Y changes for each unit increase in X

  * To make predictions
    * Once the relationship is modeled, SLR can be used to predict the value of Y for a given value of X

  * To explain variation
    * It explains how much of the variation in the dependent variable can be attributed to the independent variable.

2. What are the key assumptions of Simple Linear Regression?
- The key assumptions of Simple Linear Regression (SLR) ensure that the model is valid and that statistical inferences (like hypothesis tests and confidence intervals) are reliable

- Linearity
  * The relationship between the independent variable X and the dependent variable Y is linear
    * The mean of Y changes linearly with X

- Independence of Errors
  * The error terms (Îµ) are independent of each other  
    * No correlation between residuals (important in time-series data)

- Homoscedasticity (Constant Variance)
  * The variance of the error terms is constant across all values of X.
    * Residuals should have equal spread along the regression line.

- Normality of Errors
  * The error terms are normally distributed with a mean of zero.
    * Especially important for hypothesis testing and confidence intervals.    

- Zero Mean of Errors
  * The expected value of the error term is zero:    
  * E(Îµ)=0
    * Ensures unbiased estimates of regression coefficients.

- No Perfect Multicollinearity
  * Since SLR has only one predictor, this assumption is automatically satisfied.

9. Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.  


```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])

# Fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict values
y_pred = model.predict(X)

# Plot data points and regression line
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

10.  How do you interpret the coefficients in a simple linear regression model?

- In a simple linear regression model, the equation is:
  * y=Î²0â€‹+Î²1â€‹x+Îµ
  * The coefficients (Î²0and ğ›½1) have specific interpretations:

- Intercept (ğ›½0)
  * Represents the expected value of y when the independent variable x=0.
  * It is the point where the regression line crosses the y-axis.
  * Example: If Î²0=5, then when x=0, the predicted y=5.

- Slope (ğ›½1)  
  * Represents the expected change in y for a one-unit increase in x.
  * Determines the direction and steepness of the line:
  * Positive slope (Î²1>0): y increases as x increases.
  * Negative slope (Î²1< 0): y decreases as x increases.
  * Example: If Î²1=2, then for every 1 unit increase in x, y is expected to increase by 2 units.

- Putting it together  
  * The regression line shows the relationship between x and y.
  * The slope tells you how sensitive y is to changes in x.
  * The intercept provides a baseline value of y when x=0.
